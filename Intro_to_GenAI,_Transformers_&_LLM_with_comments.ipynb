# **1. Text Generation**

## Text Generation using GPT-2 Model

- Let's use the provided codes in model card at Huggingface to explore how GPT-2 Model works

- You can use this model directly with a pipeline for text generation. Since the generation relies on some randomness, we set a seed for reproducibility:


```python
from transformers import pipeline, set_seed  ##pipeline is a simple interface to use pretrained models.
## It automatically handles:loading the model, processing the input,  generating the output
## So instead of writing many lines of code, we can do everything in one line.

generator = pipeline('text-generation', model='gpt2') # creates a pipeline which a. generated text using b. model gpt2
# so internally this line downloads the respective model , loads into memory and prepares it to generate text

set_seed(40) # fixing the randomness, because text generation is probabilistic, the output may change each time.

generator("Hello, I'm a language model,", max_length=30, num_return_sequences=3) # sets the maximum number of words/tokens in the generated text.
# so the output will stop after 30 words/ tokens  # num sequence tells number of outputs

```

In the above code we might have doubt that why outputs are so big even if max length was provided as 30 , reason below :

In  recent versions of Hugging Face Transformers, the text-generation pipeline has default generation parameters.

For models like GPT-2, the default is often:

max_new_tokens = 256

So internally the pipeline behaves like:

generator(
    "Hello, I'm a language model,",
    max_length=30,
    max_new_tokens=256
)

When both exist, max_new_tokens wins.

1️⃣ Why randomness exists in text generation

Models like GPT-2 do not always choose the single most likely next word.
Instead, they often sample from several possible words.

Example prompt:

The weather today is

Possible next words with probabilities:

Word	Probability
sunny	0.40
cloudy	0.30
rainy	0.20

Because the model samples from these probabilities, each run may produce different text.

Example outputs:

The weather today is sunny and pleasant.
The weather today is cloudy with light winds.


2️⃣ What set_seed(42) does

set_seed(42) fixes the random number generator.

That means the random choices made during generation will be exactly the same every time.

So if you run the code again, you will get the same generated sentences.

This function comes from Hugging Face Transformers.

# **2. Text Summarization & Translation**

#### **SentencePiece Libray:**

- A powerful and efficient text processing library for NLP tasks. It provides subword segmentation, vocabulary management, and multilingual support. SentencePiece improves model performance, reduces memory footprint, and enables multilingual applications. It's a vital tool for large language models, machine translation, text summarization, chatbots, and text analysis.


```python
!pip install sentencepiece

import sentencepiece

print(sentencepiece.__version__)
```

2️⃣ Why SentencePiece is needed

Language models cannot understand raw text directly.

They first convert text into tokens (numbers).

Example:

Hello world

might become

[15496, 995]

This process is called tokenization.

SentencePiece is a tokenizer developed by Google that converts text into tokens.

3️⃣ Why normal word splitting is not enough

If we simply split by spaces:

I love playing football

we get

["I", "love", "playing", "football"]

But this causes problems because:

unknown words appear

spelling variations exist

languages like Japanese have no spaces

SentencePiece solves this by breaking words into subwords.

Example:

playing → play + ing
unbelievable → un + believe + able

So even if the model never saw “unbelievable”, it can still understand the pieces.

4️⃣ Why Hugging Face sometimes needs SentencePiece

Some transformer models use SentencePiece tokenization.

Examples include:

T5

ALBERT

mT5

If you try to load those models without installing SentencePiece, the tokenizer will fail.

So tutorials often install it first.

5️⃣ Why GPT-2 usually doesn’t need it

The model GPT-2 uses a different tokenizer called Byte-Pair Encoding (BPE).

So GPT-2 does not require SentencePiece.

But many other models do.

#### Flat-T5
- T5, short for "Text-to-Text Transfer Transformer," is a powerful Large Language Model (LLM) developed by Google AI. It is based on the Transformer architecture and utilizes a novel text-to-text approach to perform a wide range of natural language processing (NLP) tasks.

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the pretrained tokenizer for the FLAN-T5 model
# The tokenizer converts text into tokens (numbers) that the model understands
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")

# Load the pretrained FLAN-T5 model used for text generation tasks
# This model can perform tasks like translation, summarization, and question answering
# As discussed you can see more details from hugging face
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

# Define the input text with an instruction for the model
# Models like FLAN-T5 were trained using instruction tuning. During training they saw examples like:
# translate English to German: I love you → Ich liebe dich, summarize: [paragraph] → [summary], etc
# So the model learns to treat the first part as a task description.
input_text = "translate English to German: How old are you?"
print("input text:", input_text)

# Convert the input text into token IDs (numerical representation)
# return_tensors="pt" means the output will be returned as a PyTorch tensor
input_ids = tokenizer(input_text, return_tensors="pt").input_ids
print("input_ids", input_ids)

outputs = model.generate(input_ids)
print("output: ", outputs)

print(tokenizer.decode(outputs[0]))
```

In above output : "<pad> Wie old sind Sie?</s>" <pad>, </s> are special tokens which generally comes in flant5 model output, hence we can skip these through syntax as shown in below blocks

```python
input_text = "translate English to French: I love learning artificial intelligence."
input_ids = tokenizer(input_text, return_tensors="pt").input_ids

outputs = model.generate(input_ids)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

### Summarize the conversation

##### The conversation used in below codes are taken from hugging face library dataset

- Dataset Name: [knkarthick/dialogsum ](https://huggingface.co/datasets/knkarthick/dialogsum/viewer/default/train)

```python
# Let's summarize the conversation

# Define the text to summarize
input_conversation = '''
#Person1#: Look! This picture of Mom in her cap and gown.
#Person2#: Isn't it lovely! That's when she got her Master's Degree from Miami University.
#Person1#: Yes, we are very proud of her.
#Person2#: Oh, that's a nice one of all of you together. Do you have the negative? May I have a copy?
#Person1#: Surely, I'll have one made for you. You want a print? #Person2#: No. I'd like a slide, I have a new projector.
#Person1#: I'd like to see that myself. #Person2#: Have a wallet size print made for me, too. #Person1#: Certainly.
'''

Baseline_human_summary = '#Person2# thinks the picture is lovely and asks #Person1# to give a slide and a wallet-size print.'

# Prepend the summarization prefix
input_conversation = "summarize: " + input_conversation  ## same instruction format for flant5

# Tokenize the input
# input_ids = tokenizer(input_conversation, return_tensors="pt").input_ids  ## previous version of code
inputs = tokenizer(
    input_conversation,
    return_tensors="pt",
    truncation=True,
    max_length=512
)

# Generate the summary
# outputs = model.generate(input_ids, max_new_tokens=30)
outputs = model.generate(**inputs, max_new_tokens=30)
Model_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Print the summary
print('-' * 100)
print('Baseline human summary:', Baseline_human_summary)
print('-' * 100)
print("Model Summary:", Model_summary)
```

```python
# Let's summarize the conversation

# Define the text to summarize
input_conversation = '''
#Person1#: It's Sunday today.
#Person2#: Yes, I know.
#Person1#: I think we should have a house cleaning today. What's your opinion?
#Person2#: Oh, no. We just did it last week.
#Person1#: Come on. What do you want to do? Washing clothes or cleaning the house?
#Person2#: I'd rather wash the clothes.
#Person1#: Okay. Here is the laundry.
#Person2#: Oh, My God! So much!
#Person1#: Don't worry. I'll help you with it later.
'''

Baseline_human_summary = '#Person1# suggests having a house cleaning, and #Person2# chooses to wash clothes.'

# Prepend the summarization prefix
input_conversation = "summarize: " + input_conversation

# Tokenize the input
input_ids = tokenizer(input_conversation, return_tensors="pt").input_ids

# Generate the summary
outputs = model.generate(input_ids)
Model_summary = tokenizer.decode(outputs[0])

# Print the summary
print('-' * 100)
print('Baseline human summary:', Baseline_human_summary)
print('-' * 100)
print("Model Summary:", Model_summary)
```
